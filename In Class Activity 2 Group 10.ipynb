### Defining Terms
- Temporal: Type of coding based on the timing of the spikes
- Rate: Type of coding based on the frequency/average firing rate of spikes
- Latency: Type of coding based on the first spike after stimulus
- Synchronous: Type of coding based on the timing of spikes across different neurons
- Tuning Curves: Graphical representation of how neurons respond to different stimuli
- Population Codes: Information generated by combining information from multiple neurons
